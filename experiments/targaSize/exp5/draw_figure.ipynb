```python
import csv
import skimage
from skimage import data
from numpy import array
import matplotlib.pyplot as plt
import numpy as np
```

```python
csv.register_dialect('myDialect',
delimiter = ',',
skipinitialspace=True)
images=[]
labels=[]
testImages=[]
testLabels=[]
trainShape_H=[]
testShape_H=[]
trainShape_W=[]
testShape_W=[]
with open('trainVal.csv','r') as csvFile:
    reader = csv.reader(csvFile,dialect='myDialect')
    for row in reader:
        if row[3]=='1':
            img=skimage.data.imread(row[1])
            #img = skimage.transform.resize(img,(40,200,3))
            
            
            labels.append(str(row[2]))
            trainShape_H.append(img.shape[0])
            trainShape_W.append(img.shape[1])
        if row[3]=='0':
            img=skimage.data.imread(row[1])
            #img = skimage.transform.resize(img,(40,200,3))
            testShape_H.append(img.shape[0])
            testShape_W.append(img.shape[1])
            
            #testImages.append(img)
            testLabels.append(str(row[2]))
        
        #print(row,row[3])
        
```

```python
plt.style.use('seaborn-deep')
plt.axis([30,60,0,25000])
plt.xlabel('width')
plt.ylabel('number of images')
plt.title('image size comparsion on dataset')
bins = np.linspace(30, 60, 20)
plt.hist([trainShape_H,testShape_H],bins,label=['train', 'test'] )
#plt.hist(testShape_H,range=(25,65), normed=False)
plt.legend(loc='upper right')
```
